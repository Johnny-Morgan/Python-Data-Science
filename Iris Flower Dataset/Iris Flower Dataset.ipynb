# 2-D Scatter Plot

```python
iris.plot(kind = "scatter", x = "SepalLengthCm", y = "SepalWidthCm" )
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue = "Species", height = 5) \
.map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
.add_legend()
plt.show()
```

###### Observations:
1. Using SepalLengthCm and SepalWidthCm features, we can distinguish Setosa flower from others.
2. Separating Versicolor from Virginica is much harder as they have considerable overlap.


# Pair-plot

```python
sns.set_style("whitegrid")
sns.pairplot(iris, hue = "Species", height = 5)
plt.show()
```

###### Obervations:
1. PetalLengthCm and PetalWidthCm are the most useful features to identify various flower types.
2. While Setosa can be easily identified (linearly seperable), Virginica and Versicolor have some overlap (almost linearly seperable).
3. We can find "lines" and "if-else" conditions to build a simple model to classify the flower types.
